```python
# Function

# 더하기 계산기 만들기


def awe_sum(num1,num2):
    # 안전하게 하기 위해서는
    # result를 초기화 해주는 과정을 미리 포함하면 된다.
    result = 0
    result = num1 + num2
    return result


print (awe_sum(1,2))
    
```

```python
# return skill >>> return이 있고 없고의 차이를 확실히 볼 수 있는 함수 

def id_check(id):
    if id == 'admin':
        print ('Invalid ID: admin')
        return      # return이 없으면 admin 입력시 if 문 출력하고 이후 (Valid ID까지 출력함)
    print ("Valid ID: ", id)
    
id = input("what id do you want to have? : ")

print (id_check(id))
```

```python
# 사용자가 입력하지 않으면 기본값을 가지는 함수 만들기
#parameter with initialize

def say_hello(name='Fool', nick = True):   #def say_hello(name,nick) 에서 초기값을 주는 것.
    print ("Hi, ", name)
    if nick == True:                       # 사용자가 입력하지 않으면 Fool과 Ture를 가지고 출력함.
        return ("You are Fool.")            # 아무것도 입력하지 않으면 기본값이 자동적으로 입력됨.
    else:
        return ("You are not Fool.")
        
print(say_hello())
print(say_hello('name'))
print(say_hello('name', 'nickname'))

# 초기값을 설정할때는 항상 그 인자를 마지막에 두어야 합니다. 
# (name = 'Fool, nick)  >> 불가능
# (name, nick = True)   >> 가능


#기본값을 주는 것이 좋은 것은 >> 내가 원하는 방향으로 사용자의 입력을 이끌어 나갈 수 있다. 
```

```python
#keyword arguments 01

# >> *(에스터리스크)를 달면 parameter가 여러개가 들어올 수 있다.
# def function(*parameter) >>>> 여러개의 parameter를 받을 수 있다.

def mul_sum(*a):
    result = 0
    print (a)  # >>>>> *parameter은 튜플 형태로 받는다. 
    for i in a:
        result += i
    return result

print (mul_sum(20, 10, 20, 20, 30))
```

```python
#Keyword arguments 02

#별이 하나 붙었을때, 두개 붙었을때 모두 keyword arguments.
#단지 둘은 쓰임새가 다를뿐.
#두개가 붙으면 dictionary 형태로 저장됨.

def show_kwargs(**kwargs):
    print (kwargs)
    
print (show_kwargs(a=10,b="google"))
```

```python
# url완성 시키는 프로그램 만들어보기
# https 로 시작해서 서버(도메인) + 포트넘버 + ? +
# 쿼리들을 분류해서 key와 value를 찾고 위에 붙여줌. 

def kwargs_url(server, port, **query):   #쿼리 뜻: 질문, 질의어
    url = "https://" + server + ":" + port + "?"   # 도메인과 포트를 구분짓는 것이 콜론,,, 쿼리는 물음표부터 시작.>> 
    # 물음표 뒤에서 내가 어떤 브라우저를 쓰는지, 어떤 검색을 할 건지 이어짐. 
    #쿼리들을 key와 value를 = 으로 만들고 url에 더해줌
    for key in query.keys():  #쿼리라는 딕셔너리에서 키값들을 다 불러옴
        url = url + key + "=" + query[key] + "&"  # 여러개의 쿼리를 쓰기에 ampersand 필요!
        #완벽한 url 나옴.
    return url
    
print (kwargs_url("www.naver.com", "8080", ie="utf-8", query = "덩케르크"))


'''
크롤러들의 url를 완성시킬때 쓰는 함수

1. ?가 나오기 전까지 url은 정해져있는 형태

2. 쿼리라는 dictionary를 이용해서 key만 뽑아서 리스트를 만들고 반복문을 만듬.

3. 뽑아내는 키를 이용해 해당하는 밸류를 얻고 이 둘을 url에 붙여줌. 


# 쿼리가 여러개이고 쿼리의 형태는 a = something 이라는 dictionary 형태이기에 **(별표두개)를 이용해서 input을 받는것.
'''
```

```python
# keyword arguement를 사용할때는 확장가능한 값(**로 표현할)은 가장 마지막에 받아야 한다.
# (a, **query, b) >>> 이렇게 표현하면 오류 발생!
# (a, b, **query) >>> 이렇게 받아야 한다.
```

```python
# leap year 만들기

def is_leap(year):
    return year % 400 == 0 or (year % 4 == 0 and not year % 100 == 0)   #조건문을 ()parenthese를 이용해서 묶어 줄 수 있다. 


year = input("is leap year?: ")

print( "[", year," = leap year]", "is ", str(is_leap(year))
#-------------------------------------------------------------------------------위의 식은 잘못됨--이유 찾아야. 1900입력시 True-------

# 윤년인지 아닌지를 구분하는것.
# 대부분의 year들은 윤년이 아니다.
# 따라서 특이케이스인 윤년을 거르는 조건을 찾는게 중요
# 가장 특이한 소수 케이스는 400으로 나눴을 때 나오는 윤년.(4로 나눠지고 100으로도 나눠지는데 400으로도 나눠짐)
# 400으로 나눠지는 것을 제외하고 나면(or statement를 이용해서 제외) 4로 나눠지는데 100으로는 나눠지면 안는 것이 윤년
# '4로 나눠지는데 100으로 나눠지지 않는 것'을 제외하고 나면 윤년은. 없다.
# 다시 윤년의 조건으로 돌아가보면
# 가장 큰 조건부터 세부 조건 순서로(3)
# 1. 4로 나눠짐
# 2. 100으로 나눠지면 안됨. 
# 3. 400으로 나눠짐

# 3번 조건이 충족되면 무조건 윤년(필요충분조건)
# 2번 조건이 충족되면 1번 조건이 충족되야 윤년.(필요 조건)
# 1번 조건이 충족되면 2번 조건이 충족되야 윤년.(필요 조건)
'''
# ****** 집합 그래프를 그려보는 것도 도움이 됨
# 100으로 나눠 떨어지는 집합 a
# 4로 나눠 떨어지는 집합 b
# 400으로 나눠 떨어지는 집합 c
# a 안에 b가 있음.
# b 안에 c가 있음. 
'''


```

```python
# 리컬시브 (재귀호출)
# recursive
# recurse_beast(a-1)

times = int(input("How many times want to curse the beast??: "))
def recurse_beast(a):
    if a == 0:
        print ("Curse Complete!")
    else:
        print("Fusion!!!(%d times left)" % (a-1))
        recurse_beast(a-1) # **** 이게 핵심. 자기 자신을 기존의 매개변수-1인 input으로 다시 호출!!! >>> 그러면 여기서 부터 다시 procedure시작.
        
recurse_beast(times)
```

```python
times = int(input("How many times want to curse the beast??: "))
def recurse_beast(a):
    while a != 0:
        print("Fusion!!!(%d times left)" % (a-1))
        a -= 1 # **** 이게 핵심. 자기 자신을 기존의 매개변수-1인 input으로 다시 호출!!! >>> 그러면 여기서 부터 다시 procedure시작.
    return ("Curse Complete!")    
recurse_beast(times)
```
